```python
from scipy.integrate import quad
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats

def lamb(t):
    return 1/(t+1)

def m(t):
    return np.log(t+1)
T = 10

M , _ = quad(lamb, 0, T)
# dystrybuanta wynosi F(t)=log(t+1)/M
# odwrotna dystrybuanta wynosi t=exp(Y*M) - 1
N_array=np.zeros(1000)
for i in range(1000):
    N = np.random.poisson(M)
    t = [0]
    for i in range(N):
        u = np.random.rand()
        t.append(np.exp(u*M) - 1)
    t = sorted(t)
    N_array[i]=len(t)
t2=np.linspace(0,10,1000)
sns.ecdfplot(N_array,label="empiryczna")
plt.plot(t2,scipy.stats.poisson.cdf(t2,m(10)),label="teoretyczna")


# plt.step(t, np.arange(0, len(t), 1))
plt.xlabel('czas')
plt.ylabel('skoki')
plt.title('Niejednorodny proces poissona dla lamda=1/(t+1)')
plt.legend()
plt.show()
```
