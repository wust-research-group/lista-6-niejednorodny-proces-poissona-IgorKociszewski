```python
from scipy.integrate import quad
import numpy as np
import matplotlib.pyplot as plt


def lamb(t):
    return 1 / (t + 1)


def m(t):
    return np.log(t + 1)


T = 10

M, _ = quad(lamb, 0, T)
# dystrybuanta wynosi F(t)=log(t+1)/M
# odwrotna dystrybuanta wynosi t=exp(F*M) - 1
N = np.random.poisson(M)

# Generate arrival times
t = []
for i in range(N):
    u = np.random.rand()
    t.append(np.exp(u * M) - 1)

t = sorted(t)
print(t)

# Plotting
plt.step(t, np.arange(0, len(t), 1))
plt.xlabel("Time")
plt.ylabel("Number of arrivals")
plt.title("Simulated Poisson Process")
plt.show()
```
